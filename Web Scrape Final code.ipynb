```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_name(soup):
    
    try:
        name = soup.find("span", attrs = {'id':'productTitle'}).text.strip()
    except AttributeError:
        print('Cant extract')
        
    return name

def get_price(soup):
    
    try:
        price = soup.find("span", attrs = {'class':'a-size-medium a-color-price'}).text.strip()
    except AttributeError:
        print('Cant extract')
    
    return price

def get_rating(soup):
    
    try:
        rating = soup.find("span", attrs = {'class':'a-icon-alt'}).text
        
    except AttributeError:
        print('Cant extract')
        
    return rating
    
```

```python
if __name__ == "__main__":
    
    url = 'https://www.amazon.com/s?k=cosrx&crid=22KT3TLMCFTBI&sprefix=cosr%2Caps%2C532&ref=nb_sb_noss_2'
    headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 Edg/112.0.1722.58', 'Accept-Language': 'en-US, en;q = 0.5'})
    
    webpage = requests.get(url, headers = headers)
    soup = BeautifulSoup(webpage.content, "html.parser")
    
    links = soup.find_all("a", attrs = {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
    
    page_links = list()
    
    for link in links:
        page_links.append(link.get('href'))
    
    d = {'name' : [], 'price': [], 'rating': []}
    
    for link in page_links:
        webpage = requests.get('https://www.amazon.com/' + link, headers = headers)
        new_page_link = BeautifulSoup(webpage.content, "html.parser")
        
        d['name'].append(get_name(new_page_link))
        d['price'].append(get_price(new_page_link))
        d['rating'].append(get_rating(new_page_link))
    
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df
    amazon_df.to_csv('amazon_data.csv', index=False)


        
        
```
